**Masukan ke dalam satu fungsi**

```python
def clean_data(text,column_name):
    df['content_clean'] = df[column_name].str.lower()
    df['content_clean'] = df['content_clean'].str.replace('[^\w\s]','',regex=True)
    return df
```

**Penggunaan library Spacy untuk text processing**

```python
#import library spacy
import spacy
```

```python
#load model spacy untuk text processing
nlp = spacy.load('en_core_web_sm')
phrase = df.loc[1, 'content_clean']
phrase
```

```python
doc = nlp(phrase)
doc
```

**Tokenization**

```python
[token for token in doc]
```

**Lemmatization**

```python
[token.lemma_ for token in doc if not token.is_stop]
```
