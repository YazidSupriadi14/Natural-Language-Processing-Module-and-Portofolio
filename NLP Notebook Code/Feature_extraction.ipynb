# **Bag of Words**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
docs = [
    "kopi ini enak",
    "kopi ini tidak enak",
    "kopi mahal banget"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)

dtm_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
dtm_df.index = ["R1", "R2", "R3"]
dtm_df
```

```python
term_freq = dtm_df.sum(axis=0)
term_freq.sort_values().plot(kind="barh")
```

# **TF-IDF**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [
    "kopi ini enak",
    "kopi ini tidak enak",
    "kopi mahal"
]

tfidf = TfidfVectorizer()  # non-normalized biar lebih jelas
X = tfidf.fit_transform(docs)

# Vocabulary
vocab = tfidf.get_feature_names_out()

# IDF
idf = tfidf.idf_

# TF-IDF Matrix
tfidf_matrix = X.toarray()

print("Vocabulary:", vocab)

print("\nIDF:")
for word, val in zip(vocab, idf):
    print(f"{word}: {round(val,3)}")

print("\nTF-IDF Matrix:")
print(tfidf_matrix)
```
